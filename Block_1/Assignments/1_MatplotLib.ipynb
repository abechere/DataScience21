```python
#import an read big-mart data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Read data into pandas table - details on pandas in block 3
data = pd.read_csv("../../DATA/company_sales_data.csv") 
```

## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit

### The result should look something like this:
<img src="IMG/1.png" width=400>

```python
plt.figure()
fig, ax = plt.subplots(figsize=(6,4))
plt.plot(data.month_number, data.total_profit, 'b')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
ax.set_yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


### The result should look something like this:
<img src="IMG/2.png" width=400>

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(data.month_number, data.total_units, color="red", lw=3, ls='--', marker='o', 
        markerfacecolor="blue", label="Profit data of last year")

ax.set_xlabel('Month Number')
ax.set_ylabel('Sold units number')
ax.set_title('Company Sales data of last year');

ax.set_yticks(range(10000, 60000, 10000))
ax.set_xticks(range(1,13))
ax.legend(loc=4)
plt.show()
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(data.month_number, data.facecream, color="blue", lw=3, ls='-', marker='o', label="Face cream")
ax.plot(data.month_number, data.facewash, color="orange", lw=3, ls='-', marker='o', label="Face Wash")
ax.plot(data.month_number, data.toothpaste, color="green", lw=3, ls='-', marker='o', label="Toothpaste")
ax.plot(data.month_number, data.bathingsoap, color="red", lw=3, ls='-', marker='o', label="Bathing Soap")
ax.plot(data.month_number, data.shampoo, color="violet", lw=3, ls='-', marker='o', label="Shampoo")
ax.plot(data.month_number, data.moisturizer, color="brown", lw=3, ls='-', marker='o', label="Moisturizer")

ax.set_xlabel('Month Number')
ax.set_ylabel('Sold units in number')
ax.set_title('Sales data');

ax.set_yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
ax.set_xticks(range(1,13))
ax.legend(loc=2)
plt.show()
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.


### The result should look something like this:
<img src="IMG/4.png" width=400>

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(data.month_number, data.toothpaste, 'co', label="Tooth paste Sales data")

ax.set_xlabel('Month Number')
ax.set_ylabel('Number of units Sold')
ax.set_title('Tooth paste Sales data');

ax.set_xticks(range(1,13))
ax.grid(linestyle='dashed', linewidth=0.9)
ax.legend(loc=2)
plt.show()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
fig, ax = plt.subplots(figsize=(6,4))

width = 0.3
ax.bar(data.month_number, data.facecream, color="blue", label="Face Cream", align="center", width=width, alpha=0.7)
ax.bar(data.month_number+width, data.facewash, color="orange", label="Face Wash", align="center", width=+width, alpha=0.7)

ax.set_xlabel('Month Number')
ax.set_ylabel('Sales units in number')
ax.set_title('Facewash and facecream sales data');

ax.set_xticks(range(1,13))
ax.grid(linestyle='dashed', linewidth=0.9)
ax.legend(loc=2)

plt.show()
```
